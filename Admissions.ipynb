```
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn 
from matplotlib import gridspec
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
```

```
df = pd.read_csv("/content/drive/My Drive/WGU/Admission_Predict_Ver1.1.csv")
```

```
df.head(10)
```

```
features=df.iloc[:,0:9].columns

sns.set_palette("bright")

plt.figure(figsize=(10,9*4))
gridspecs=gridspec.GridSpec(9,1)
for i, c in enumerate(df[features]):
  ax=plt.subplot(gridspecs[i])
  sns.distplot(df[c][df.Research==1])
  sns.distplot(df[c][df.Research==0])
plt.show()
```

```
sns.pairplot(df)
```

```
corr=df.corr()
sns.heatmap(corr, xticklabels=corr.columns.values,yticklabels=corr.columns.values)
```

```
X = df.drop('Chance of Admit ', axis=1)
y = df['Chance of Admit ']
#Create the seperate test and validation sets
X_train, X_Val, y_train, y_val = train_test_split(X,y,test_size=.33, random_state=111)
```

```
mlpreg = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
mlpreg.predict(X_Val[:2])

mlpreg.score(X_Val, y_val)
```
